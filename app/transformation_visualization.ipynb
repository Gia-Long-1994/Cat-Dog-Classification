```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import tensorflow as tf
import matplotlib.pyplot as plt
```

## Transformation list
- **rotation_range=90** *randomly rotate image with given range*
- **width_shift_range=0.2** *randomly shift image vertically*
- **height_shift_range=0.2** *randomly shift image horizontally*
- **shear_range=0.2** *randomly applying shearing transformations*
- **zoom_range=0.2** *randomly zooming inside image*
- **horizontal_flip=True** *randomly flipping half of the images horizontally*

```python
image_size=300
batch_size=1
train_path='./data/rotate_test'
```

```python
image = tf.io.read_file("dog.1278.jpg")
image = tf.image.decode_jpeg(image, channels=3)
image = tf.image.resize(image, [image_size, image_size])
image /= 255.0
```

## Rotation

```python
im_gen = ImageDataGenerator(rotation_range=90)
train_gen = im_gen.flow_from_directory(directory=train_path, 
                                        batch_size=batch_size, 
                                        shuffle=True,
                                        target_size=(image_size, image_size),
                                        class_mode='binary',
                                        seed = 101)
```

```python
fig , ax = plt.subplots(1,2,figsize=(12,15),subplot_kw=dict(xticks=[], yticks=[]))
ax[0].set_title('original image',fontsize=20)
ax[1].set_title('rotated image',fontsize=20)
ax[0].imshow(image)
ax[1].imshow(train_gen[0][0][0]/255)
```

## Width shift

```python
im_gen = ImageDataGenerator(width_shift_range=0.6)
train_gen = im_gen.flow_from_directory(directory=train_path, 
                                        batch_size=batch_size, 
                                        shuffle=True,
                                        target_size=(image_size, image_size),
                                        class_mode='binary',
                                        seed = 101)
```

```python
fig , ax = plt.subplots(1,2,figsize=(12,15),subplot_kw=dict(xticks=[], yticks=[]))
ax[0].set_title('original image',fontsize=20)
ax[1].set_title('width-shifted image',fontsize=20)
ax[0].imshow(image)
ax[1].imshow(train_gen[0][0][0]/255)
```
